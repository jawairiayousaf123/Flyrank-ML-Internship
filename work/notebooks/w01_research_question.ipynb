## 3. Quick look at the data (2-3 real numbers)

The starter dataset contains **30,000 pages** and **44 features**, providing a large and diverse dataset for analysis. Out of these, **16,262 pages (54.2%)** are labeled as declining, which shows that identifying pages needing attention is a meaningful problem. The dataset also has an **average CTR of 0.511** and an **average of 5,200.37 impressions** over 90 days, indicating that page performance varies and can be analyzed to support refresh decisions.

These numbers suggest that a data-driven approach can help prioritize which pages should be reviewed first instead of relying only on manual inspection.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print("Total Pages:", len(df))
print("Total Columns:", len(df.columns))
print("Declining Pages:", (df["trend_direction"]=="down").sum())
print("Declining Rate:", round((df["trend_direction"]=="down").mean(),3))
print("Average CTR:", round(df["ctr"].mean(),3))
print("Average Impressions:", round(df["impressions_90d"].mean(),2))
```
