## 2. Target or proxy

The target (proxy) is **is_declining_label**.

This label is derived from the **trend_direction** column, where:

- Down = 1 (Declining)
- Up or Stable = 0 (Not Declining)

This target allows us to train a model that identifies pages that may require a content refresh.



```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import os
import pandas as pd

os.chdir("/content/flyrank-ml-internship-starter")

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

df["is_declining_label"] = (
    df["trend_direction"].str.lower() == "down"
).astype(int)

df[["trend_direction", "is_declining_label"]].head()
```

## 3. Success metric

The primary success metric is Precision@50.

Precision@50 measures how many of the top 50 recommended pages are actually declining.

A higher Precision@50 means the content team spends more time updating pages that truly need attention instead of reviewing low-priority pages.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
print("Evaluation Metric: Precision@50")
```

## 5. Why ML beats a fixed rule here

A fixed rule relies on only one or two manually selected conditions, such as pages that have not been updated for many days.

Machine learning can combine many different signals at the same time, including impressions, CTR, average position, content age, and update history.

This makes the recommendations more flexible and data-driven while supporting better prioritization decisions.
